```python
# Therefore, I am solving assignment number two using the code I have created in Task 1.
#            -> Here I have worked with a standard label enconding technique
#            -----(see source:https://pbpython.com/categorical-encoding.html)-----
#            -> I will be adding two new columns to the data frame.
#            -> The first new dummy variable is called: "genderencoding" which diffentiates man, woman and non-natural persons;
#               And it follows the following transformation:
#                   1 == male; Natural Person attribute
#                   0 == female; Natural Person attribute
#                   NaN == no Gender information and therefore a company or "non-natural person"
#            -> The second (per your request): "gender_manual" just differentiating: naturalperson & all_gender.
#               And it follows the following transformation:
#                    1 == natural_person; Natural Person attribute
#                    NaN == all_gender; Natural Person attribute
```

```python
# Checking df data types and doing some small shape changes

df.dtypes # All dtypes are objects -> I want categories
df = df.rename(columns=df.iloc[0]).drop(labels=0,axis="index") # Renaming header, and saving current shape & Droping index 0
```

```python
df["gender"] = df.gender.astype("category")  # Transforming gender data type into "categorical"
df.dtypes # Checking if it worked -> It worked.
```

```python
# <DUMMY VARIABLE: "genderencoding">
df["gender"].value_counts() # Checking categorical variables under column Gender
```

```python
df["genderencoding"] = df["gender"].cat.codes # Creating a copy of column "Gender" and adding it to the data frame, in order to apply the label 
                                              # enconding. The encoding is almost done, the only issue is: NaN values received the numeric value 
                                              # "-1" and I want them to be empty.
```

```python
df["genderencoding"] = df["genderencoding"].replace(to_replace=-1, value="NaN") # Done! NaN values under "GenderEnconding" that were 
df.tail()                                                                       # encoded to "-1" are now the standard "NaN" again.
```

```python
# <Dummy Variable: gender_manual> -> Simplified version of "genderencoding" differetiating only natural from non-natural person
df["gender_manual"] = df["genderencoding"] # Created new dummy variable
df.tail()
```

```python
df["gender_manual"] = df["gender_manual"].replace(to_replace=0, value="1") # Shaped the dummy variable to differetiate natural persons with a numeric value = 1 
#                                                                            & non-natural persons with a numeric value = NaN. Assignment 2 is done. 
df.head() # Checking if it works and it does.
```

```python
# <TASK 1: Making dummy variable more robust>  Previosly, I have filtered it just using Gender because 
#                                              I assumed the data was 100% reliable - amateur mistake sorry. 

df = df.drop("naturalperson", axis=1) # Deleting old dummy variable "naturalperson" 
```

```python
# Now I am applying a custom binary enconding technique. 
#     Here I am looking for specific string values in different columns. 
#     If there is a string value I am looking for dummy variable: 
#     "NaturalPersonTitleCheckStatus" has a value of 1 otherwise 0.
#     My objective is at the end to come up with a score regarding: the
#     likelihood of having a natural person based on this attributes:
#             1. String value (natural person) in col "dmfullname" (40%)
#             2. Strinf value (campany titles & signs& numbers) (40%)
#             4. Gender_manual (20%)

df["SirTitleCheck"] = np.where(df["dmfullname"].str.contains("Sir"), 1, 0)                   # Checking string values regarding natural persons name. The code here is repetitive, using str.contains() I did't find a way to iterate multiple labels at the same time - sorry :/
df["MadamTitleCheck"] = np.where(df["dmfullname"].str.contains("Madam"), 1, 0)
df["MsTitleCheck"] = np.where(df["dmfullname"].str.contains("Ms"), 1, 0)
df["MrTitleCheck"] = np.where(df["dmfullname"].str.contains("Mr"), 1, 0)
df["MrsTitleCheck"] = np.where(df["dmfullname"].str.contains("Mrs"), 1, 0)
df["MissTitleCheck"] = np.where(df["dmfullname"].str.contains("Miss"), 1, 0)
df["DrTitleCheck"] = np.where(df["dmfullname"].str.contains("Dr"), 1, 0)
df["ProfessorTitleCheck"] = np.where(df["dmfullname"].str.contains("Professor"), 1, 0)

cols = ["SirTitleCheck","MadamTitleCheck", "MsTitleCheck", "MrTitleCheck", "MrsTitleCheck", "MrsTitleCheck", "MissTitleCheck", "DrTitleCheck", "ProfessorTitleCheck"]
df["NaturalPersonTitleCheck"] = df[cols].apply(lambda row: "+" .join(row.values.astype(str)), axis=1) # Joining all new columns, and having them as string dtype 

df["NaturalPersonTitleCheckStatus"] = np.where(df["NaturalPersonTitleCheck"].str.contains("1"), 1, 0) # If a 1 is part of this string, it means that a Natural person title was found. Therefore, 1 represents possible natural persons while 0 non-natural persons.
df = df.drop(["SirTitleCheck","MadamTitleCheck", "MsTitleCheck", "MrTitleCheck", "MrsTitleCheck", "MrsTitleCheck", "MissTitleCheck", "DrTitleCheck", "ProfessorTitleCheck", "NaturalPersonTitleCheck"], axis=1)
df.head()
```
