```python
import pandas as pd
import re
import numpy as np
```

```python
df = pd.read_table("Lucas_unique_directors_naturalperson_gender.csv", header= None, sep= None, engine="python")
```

```python
# <TASK 2: Gender Differentiation>

# Eline, in the first assigment of this task (jupyter script: "Lucas_naturalperson_gender_worked.ipynb"), 
#           I came up with the dummy variable, "natural person", by checking if Gender was given because a natural person 
#           has Gender while, e.g. a company, does not. This fulfills Task. 1 if the data is perfect / correct to its 
#           full potential, meaning every natural person observation has a categorical value under column Gender and every "non-natural person" not - which
#           might not be the case. However, as Task. 2 regards the differentiation of attribute Gender, I can use the code
#           for Task. 1 to solve Task. 2 and approach Task. 1 in a different way shown below.

# Therefore, I am solving assignment number two using the code I have created in Task 1.
#            -> Here I have worked with a standard label enconding technique
#            -----(see source:https://pbpython.com/categorical-encoding.html)-----
#            -> I will be adding two new columns to the data frame.
#            -> The first new dummy variable is called: "genderencoding" which diffentiates man, woman and non-natural persons;
#               And it follows the following transformation:
#                   1 == male; Natural Person attribute
#                   0 == female; Natural Person attribute
#                   NaN == no Gender information and therefore a company or "non-natural person"
#            -> The second (per your request): "gender_manual" just differentiating: naturalperson & all_gender.
#               And it follows the following transformation:
#                    1 == natural_person; Natural Person attribute
#                    NaN == all_gender; Natural Person attribute
```

```python
# Checking df data types and doing some small shape changes

df.dtypes # All dtypes are objects -> I want categories
df = df.rename(columns=df.iloc[0]).drop(labels=0,axis="index") # Renaming header, and saving current shape & Droping index 0
```

```python
df["gender"] = df.gender.astype("category")  # Transforming gender data type into "categorical"
df.dtypes # Checking if it worked -> It worked.
```

```python
# <DUMMY VARIABLE: "genderencoding">
df["gender"].value_counts() # Checking categorical variables under column Gender
```

```python
df["genderencoding"] = df["gender"].cat.codes # Creating a copy of column "Gender" and adding it to the data frame, in order to apply the label 
                                              # enconding. The encoding is almost done, the only issue is: NaN values received the numeric value 
                                              # "-1" and I want them to be empty.
```

```python
df["genderencoding"] = df["genderencoding"].replace(to_replace=-1, value="NaN") # Done! NaN values under "GenderEnconding" that were 
df.tail()    
```

```python
# <Dummy Variable: gender_manual> -> Simplified version of "genderencoding" differetiating only natural from non-natural person
df["gender_manual"] = df["genderencoding"] # Created new dummy variable
df.tail()
```

```python
df["gender_manual"] = df["gender_manual"].replace(to_replace=0, value="1") # Shaped the dummy variable to differetiate natural persons with a numeric value = 1 
#                                                                            & non-natural persons with a numeric value = NaN. Assignment 2 is done. 
df.head() # Checking if it works and it does.
```

```python
# <TASK 1: Making dummy variable more robust>  Previosly, I have filtered it just using Gender because 
#                                              I assumed the data was 100% reliable - amateur mistake sorry. 

df = df.drop("naturalperson", axis=1) # Deleting old dummy variable "naturalperson" 
```

```python
# Now I am applying a custom binary enconding technique. 
#     Here I am looking for specific string values in different columns. 
#     If there is a string value I am looking for dummy variable: 
#     "NaturalPersonTitleCheckStatus" has a value of 1 otherwise 0.


df["SirTitleCheck"] = np.where(df["dmfullname"].str.contains("Sir"), 1, 0)                   # Checking string values regarding natural persons name. The code here is repetitive, using str.contains() I did't find a way to iterate multiple labels at the same time - sorry :/
df["MadamTitleCheck"] = np.where(df["dmfullname"].str.contains("Madam"), 1, 0)
df["MsTitleCheck"] = np.where(df["dmfullname"].str.contains("Ms"), 1, 0)
df["MrTitleCheck"] = np.where(df["dmfullname"].str.contains("Mr"), 1, 0)
df["MrsTitleCheck"] = np.where(df["dmfullname"].str.contains("Mrs"), 1, 0)
df["MissTitleCheck"] = np.where(df["dmfullname"].str.contains("Miss"), 1, 0)
df["DrTitleCheck"] = np.where(df["dmfullname"].str.contains("Dr"), 1, 0)
df["ProfessorTitleCheck"] = np.where(df["dmfullname"].str.contains("Professor"), 1, 0)

cols = ["SirTitleCheck","MadamTitleCheck", "MsTitleCheck", "MrTitleCheck", "MrsTitleCheck", "MrsTitleCheck", "MissTitleCheck", "DrTitleCheck", "ProfessorTitleCheck"]
df["NaturalPersonTitleCheck"] = df[cols].apply(lambda row: "+" .join(row.values.astype(str)), axis=1) # Joining all new columns, and having them as string dtype 

df["NaturalPersonTitleCheckStatus"] = np.where(df["NaturalPersonTitleCheck"].str.contains("1"), 1, 0) # If a 1 is part of this string, it means that a Natural person title was found. Therefore, 1 represents possible natural persons while 0 non-natural persons.
df = df.drop(["SirTitleCheck","MadamTitleCheck", "MsTitleCheck", "MrTitleCheck", "MrsTitleCheck", "MrsTitleCheck", "MissTitleCheck", "DrTitleCheck", "ProfessorTitleCheck", "NaturalPersonTitleCheck"], axis=1)
df.head()
```

```python
# Now I wil apply same procedure but looking for numerical string values in dmfullname: e.g. "1", "2", "3"
     
df["Numeric9TitleCheck"] = np.where(df["dmfullname"].str.contains("9"), 1, 0) # Checking string values regarding non-natural persons name. The code here is repetitive, using str.contains() I did't find a way to iterate multiple labels at the same time - sorry :/
df["Numeric8TitleCheck"] = np.where(df["dmfullname"].str.contains("8"), 1, 0) 
df["Numeric7TitleCheck"] = np.where(df["dmfullname"].str.contains("7"), 1, 0) 
df["Numeric6TitleCheck"] = np.where(df["dmfullname"].str.contains("6"), 1, 0)
df["Numeric5TitleCheck"] = np.where(df["dmfullname"].str.contains("5"), 1, 0) 
df["Numeric4TitleCheck"] = np.where(df["dmfullname"].str.contains("4"), 1, 0) 
df["Numeric3TitleCheck"] = np.where(df["dmfullname"].str.contains("3"), 1, 0)
df["Numeric2TitleCheck"] = np.where(df["dmfullname"].str.contains("2"), 1, 0) 
df["Numeric1TitleCheck"] = np.where(df["dmfullname"].str.contains("1"), 1, 0) 
df["Numeric0TitleCheck"] = np.where(df["dmfullname"].str.contains("0"), 1, 0)

numeric_filter = ["Numeric9TitleCheck", "Numeric8TitleCheck", "Numeric7TitleCheck", "Numeric6TitleCheck","Numeric5TitleCheck", "Numeric4TitleCheck", "Numeric3TitleCheck", "Numeric2TitleCheck", "Numeric1TitleCheck", "Numeric0TitleCheck"]
df["NumericTitleCheck"] = df[numeric_filter].apply(lambda row: "+" .join(row.values.astype(str)), axis=1) # Joining all new columns, and having them as string dtype 
df["NumericTitleCheckStatus"] = np.where(df["NumericTitleCheck"].str.contains("1"), 1, 0) # If a 1 is part of this string, it means that a numeric value was found, implying this is certainly not a natural person. Therefore, 1 represents a non-natural persons while 0 natural persons.
df.drop(["Numeric9TitleCheck", "Numeric8TitleCheck", "Numeric7TitleCheck", "Numeric6TitleCheck","Numeric5TitleCheck", "Numeric4TitleCheck", "Numeric3TitleCheck", "Numeric2TitleCheck", "Numeric1TitleCheck", "Numeric0TitleCheck", "NumericTitleCheck"],axis=1)
df["NumericTitleCheckStatus"].value_counts() # Know that out of all observations 120 are non-natural person for sure
```

```python
df_demo = df.drop(["Numeric9TitleCheck", "Numeric8TitleCheck", "Numeric7TitleCheck", "Numeric6TitleCheck","Numeric5TitleCheck", "Numeric4TitleCheck", "Numeric3TitleCheck", "Numeric2TitleCheck", "Numeric1TitleCheck", "Numeric0TitleCheck", "NumericTitleCheck","S.p.a.TitleCheck", "S.r.l.TitleCheck","LtdTitleCheck", "PLCTitleCheck", "PLCTitleCheck", "B.V.TitleCheck", "LimitedTitleCheck", "GmbHTitleCheck", "AGTitleCheck", "UGTitleCheck", "e.V.TitleCheck", "SASTitleCheck", "SARLTitleCheck", "SATitleCheck", "OGTitleCheck", "KGTitleCheck", "LLCTitleCheck", "U.A.TitleCheck", "C.V.TitleCheck", "ABTitleCheck", "HBTitleCheck", "KBTitleCheck", "Zrt.TitleCheck","Kft.TitleCheck", "S.à r.l.TitleCheck", "S.A.TitleCheck", "SCSpTitleCheck", "LLCsTitleCheck", "OyTitleCheck", "LTDTitleCheck", "L.T.D.TitleCheck", "P.C.TitleCheck", "CompanyTitleCheck"],axis=1)
df_demo.head()
```

```python
# List of conditions
conditions = [
    (df_demo['NumericTitleCheckStatus'] == 1), # This conditions is secure. It describes the following: if a number exists in dmfullname -> non-natural person.
    (df_demo['CompanyTitleCheckStatus'] == 1), # This conditions is secure. It describes the following: if a company fiscal title exists in dmfullname -> non-natural person.
    (df_demo['NaturalPersonTitleCheckStatus'] == 0) & (df_demo['gender_manual'] != 1) , # This conditions is partialy secure. It describes the following: if no person title exists in dmfullname and no gender is given -> non-natural person. 
    (df_demo['NaturalPersonTitleCheckStatus'] == 1) & (df_demo['gender_manual'] != 1), # This conditions is partialy secure. Here gender has a higher weight. I am trying to address dmfullnames with natural person titles that are possible companies,e.g. Dr Schwarz, which could potentially be a company eventhough it has natural person title.
    (df_demo['NaturalPersonTitleCheckStatus'] == 1) & (df_demo['gender_manual'] == 1), # This conditions is partialy secure. It describes the following: if a person title exists in dmfullname and gender is given -> natural person.
    (df_demo['NumericTitleCheckStatus'] == 0) & (df_demo['gender_manual'] == 1), # This conditions is partialy secure. It describes the following: if no numeric value exists in dmfullname and gender is given -> natural person.
    (df_demo['CompanyTitleCheckStatus'] == 0) & (df_demo['gender_manual'] == 1), # # This conditions is partialy secure. It describes the following: if no company fiscal title exists in dmfullname and gender is given -> natural person.
]

# create a list of the values we want to assign for each condition
values = ['non', 'non', 'non', 'non','natural','natural', 'natural']

# create a new column and use np.select to assign values to it using our lists as arguments
df_demo['naturalperson'] = np.select(conditions, values)
```
